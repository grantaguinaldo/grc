```python
df_2019.shape[0]
```

```python
index_file = []
for index, row in df_2019.iterrows():
    file_path = './2019/' + row['file_name']
    
    try:
        print('{} of {} \ File Name: {}'.format(index, df_2019.shape[0], row['file_name']))
        
        pdfReader = pp.PdfFileReader(open(file_path, 'rb'))
        
        EXTRACT = [each.replace('\n', '').strip() for each in \
                 [pdfReader.getPage(each).extractText() for \
                  each in range(pdfReader.numPages)]]
        
        CLEAN_EXTRACT = [each for each in EXTRACT if each != '']
                
        df_2019.set_value(index, 'file_text', CLEAN_EXTRACT)
        
    except:
        index_file.append(row)
        print('****************************** Error on Row: {} ******************************'.format(row))

index_file
```
