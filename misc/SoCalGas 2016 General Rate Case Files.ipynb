```python
from bs4 import BeautifulSoup as soup
from pathlib import Path
import requests as r
import time

_r = r.get('https://www.socalgas.com/regulatory/A17-10-008.shtml')
_soup = soup(_r.text, 'html.parser')
links = _soup.findAll('div', {'class': 'container container-white'})
url_list = ['http://www.socalgas.com' + each['href'] for each in links[0].find_all('a', href=True)]
```

```python
for index, each in enumerate(url_list):
    print('File: {} // Name: {}'.format(index, each))
    filename = Path(each.split('/')[-1])
    response = r.get(each)
    filename.write_bytes(response.content)
    time.sleep(5)
```
